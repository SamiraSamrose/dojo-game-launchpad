```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Generate synthetic data for the launchpad ecosystem

# 1. Developer Onboarding & Game Launch Timeline
np.random.seed(42)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
devs_onboarded = [15, 28, 45, 72, 98, 125]
games_launched = [3, 8, 15, 28, 42, 61]
ai_assisted = [3, 7, 13, 25, 38, 56]

# 2. AI Agent Usage Statistics
ai_features = ['Game Publishing', 'Documentation\n(RAG)', 'Code Generation', 'Bug Fixing', 'Asset Optimization']
usage_counts = [892, 745, 623, 501, 387]

# 3. Multi-Chain Payment Distribution
payment_methods = ['Starknet\n(Chipi Pay)', 'Bitcoin\n(Xverse)', 'Bitcoin\n(Vesu)', 'Fiat Backup']
payment_volume = [45.3, 28.7, 18.2, 7.8]  # percentage
transaction_counts = [3421, 1876, 1254, 523]

# 4. Platform Component Usage
components = ['Dojo Engine', 'AI Agent', 'Payment\nGateway', 'Mobile\nLauncher', 'Privacy\nChat', 'Open\nTemplates']
daily_active = [1850, 1620, 2340, 3100, 980, 1450]

# 5. Game Template Downloads (Open Source)
templates = ['RPG\nBasic', '2D\nPlatformer', 'Card\nBattle', 'Strategy\nBase', 'Puzzle\nKit', 'Multiplayer\nStarter']
downloads = [342, 289, 256, 198, 187, 145]

# 6. Developer Workflow Stages
workflow_stages = ['Initial\nSetup', 'AI-Assisted\nDevelopment', 'Testing\n& Debug', 'Payment\nIntegration', 'Privacy\nConfig', 'Launch']
avg_time_hours = [2.3, 18.5, 12.8, 5.2, 3.1, 1.8]
success_rate = [98, 94, 89, 96, 99, 92]

# 7. Privacy Features Usage (Wootzapp)
privacy_metrics = pd.DataFrame({
    'Feature': ['Encrypted\nChat', 'Secure Key\nStorage', 'Anonymous\nAnalytics', 'Private\nTesting'],
    'Active Users': [856, 743, 921, 612],
    'Satisfaction': [4.7, 4.8, 4.5, 4.6]
})

# 8. Revenue Model Breakdown
revenue_sources = ['Transaction\nFees', 'Premium\nFeatures', 'Template\nMarket', 'API Access']
monthly_revenue = [12500, 8300, 5600, 3800]

# Create comprehensive visualization dashboard
fig = plt.figure(figsize=(20, 14))
fig.suptitle('Dojo Game Launchpad - Mobile Platform Analytics Dashboard',
             fontsize=20, fontweight='bold', y=0.995)

# 1. Developer Growth & Game Launches
ax1 = plt.subplot(3, 4, 1)
x = np.arange(len(months))
width = 0.35
bars1 = ax1.bar(x - width/2, devs_onboarded, width, label='Developers', alpha=0.8)
bars2 = ax1.bar(x + width/2, games_launched, width, label='Games Launched', alpha=0.8)
ax1.plot(x, ai_assisted, 'o-', linewidth=2, markersize=8, label='AI-Assisted', color='red')
ax1.set_xlabel('Month', fontweight='bold')
ax1.set_ylabel('Count', fontweight='bold')
ax1.set_title('Platform Growth Trajectory', fontweight='bold', pad=10)
ax1.set_xticks(x)
ax1.set_xticklabels(months)
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. AI Agent Feature Usage
ax2 = plt.subplot(3, 4, 2)
colors = plt.cm.viridis(np.linspace(0.2, 0.8, len(ai_features)))
bars = ax2.barh(ai_features, usage_counts, color=colors, alpha=0.8)
ax2.set_xlabel('Usage Count', fontweight='bold')
ax2.set_title('AI Agent Feature Utilization', fontweight='bold', pad=10)
ax2.grid(True, alpha=0.3, axis='x')
for i, v in enumerate(usage_counts):
    ax2.text(v + 20, i, str(v), va='center', fontweight='bold')

# 3. Multi-Chain Payment Distribution (Pie)
ax3 = plt.subplot(3, 4, 3)
explode = (0.05, 0.05, 0.05, 0)
colors_pie = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']
wedges, texts, autotexts = ax3.pie(payment_volume, labels=payment_methods, autopct='%1.1f%%',
                                     explode=explode, colors=colors_pie, startangle=90)
ax3.set_title('Payment Method Distribution', fontweight='bold', pad=10)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# 4. Transaction Volume by Payment Method
ax4 = plt.subplot(3, 4, 4)
colors_trans = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']
bars = ax4.bar(payment_methods, transaction_counts, color=colors_trans, alpha=0.8)
ax4.set_ylabel('Transaction Count', fontweight='bold')
ax4.set_title('Transaction Volume by Chain', fontweight='bold', pad=10)
ax4.grid(True, alpha=0.3, axis='y')
for bar in bars:
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

# 5. Platform Component Usage
ax5 = plt.subplot(3, 4, 5)
colors_comp = plt.cm.Set3(np.linspace(0, 1, len(components)))
bars = ax5.bar(components, daily_active, color=colors_comp, alpha=0.8)
ax5.set_ylabel('Daily Active Users', fontweight='bold')
ax5.set_title('Component Usage (Daily Active)', fontweight='bold', pad=10)
ax5.grid(True, alpha=0.3, axis='y')
plt.setp(ax5.xaxis.get_majorticklabels(), rotation=45, ha='right')

# 6. Open Source Template Downloads
ax6 = plt.subplot(3, 4, 6)
colors_temp = plt.cm.Pastel1(np.linspace(0, 1, len(templates)))
bars = ax6.barh(templates, downloads, color=colors_temp, alpha=0.8)
ax6.set_xlabel('Downloads', fontweight='bold')
ax6.set_title('Game Template Downloads', fontweight='bold', pad=10)
ax6.grid(True, alpha=0.3, axis='x')

# 7. Developer Workflow Timeline
ax7 = plt.subplot(3, 4, 7)
colors_workflow = plt.cm.coolwarm(np.array(success_rate)/100)
bars = ax7.bar(workflow_stages, avg_time_hours, color=colors_workflow, alpha=0.8)
ax7.set_ylabel('Average Time (hours)', fontweight='bold')
ax7.set_title('Development Workflow Duration', fontweight='bold', pad=10)
ax7.grid(True, alpha=0.3, axis='y')
plt.setp(ax7.xaxis.get_majorticklabels(), rotation=45, ha='right')

# 8. Workflow Success Rates
ax8 = plt.subplot(3, 4, 8)
ax8.plot(workflow_stages, success_rate, 'o-', linewidth=3, markersize=10,
         color='green', markerfacecolor='lightgreen', markeredgewidth=2)
ax8.fill_between(range(len(workflow_stages)), success_rate, alpha=0.3, color='green')
ax8.set_ylabel('Success Rate (%)', fontweight='bold')
ax8.set_title('Workflow Stage Success Rates', fontweight='bold', pad=10)
ax8.set_ylim([85, 100])
ax8.grid(True, alpha=0.3)
plt.setp(ax8.xaxis.get_majorticklabels(), rotation=45, ha='right')

# 9. Privacy Features Analysis
ax9 = plt.subplot(3, 4, 9)
x_privacy = np.arange(len(privacy_metrics))
bars1 = ax9.bar(x_privacy, privacy_metrics['Active Users'], alpha=0.7, label='Active Users')
ax9.set_ylabel('Active Users', fontweight='bold', color='blue')
ax9.set_xticks(x_privacy)
ax9.set_xticklabels(privacy_metrics['Feature'])
ax9.tick_params(axis='y', labelcolor='blue')
plt.setp(ax9.xaxis.get_majorticklabels(), rotation=45, ha='right')

ax9_twin = ax9.twinx()
ax9_twin.plot(x_privacy, privacy_metrics['Satisfaction'], 'ro-', linewidth=2,
              markersize=10, label='Satisfaction', markerfacecolor='yellow')
ax9_twin.set_ylabel('Satisfaction (out of 5)', fontweight='bold', color='red')
ax9_twin.set_ylim([4.0, 5.0])
ax9_twin.tick_params(axis='y', labelcolor='red')
ax9.set_title('Privacy Features (Wootzapp)', fontweight='bold', pad=10)
ax9.legend(loc='upper left')
ax9_twin.legend(loc='upper right')
ax9.grid(True, alpha=0.3)

# 10. Revenue Model Breakdown
ax10 = plt.subplot(3, 4, 10)
colors_rev = ['#FF6B6B', '#4ECDC4', '#95E1D3', '#F38181']
bars = ax10.bar(revenue_sources, monthly_revenue, color=colors_rev, alpha=0.8)
ax10.set_ylabel('Monthly Revenue ($)', fontweight='bold')
ax10.set_title('Revenue Sources', fontweight='bold', pad=10)
ax10.grid(True, alpha=0.3, axis='y')
plt.setp(ax10.xaxis.get_majorticklabels(), rotation=45, ha='right')
for bar in bars:
    height = bar.get_height()
    ax10.text(bar.get_x() + bar.get_width()/2., height,
              f'${int(height):,}', ha='center', va='bottom', fontweight='bold')

# 11. Technology Stack Heatmap
ax11 = plt.subplot(3, 4, 11)
tech_data = np.array([
    [95, 88, 72, 65],  # Dojo Engine
    [92, 94, 78, 71],  # AI Agent
    [88, 91, 96, 82],  # Payment Gateway
    [97, 89, 85, 79],  # Mobile Launcher
    [85, 99, 76, 88],  # Privacy Layer
    [91, 84, 68, 92]   # Open Templates
])
tech_labels = ['Dojo', 'AI Agent', 'Payments', 'Mobile', 'Privacy', 'Templates']
metric_labels = ['Reliability', 'Security', 'Speed', 'Adoption']
im = ax11.imshow(tech_data, cmap='YlGn', aspect='auto')
ax11.set_xticks(np.arange(len(metric_labels)))
ax11.set_yticks(np.arange(len(tech_labels)))
ax11.set_xticklabels(metric_labels)
ax11.set_yticklabels(tech_labels)
plt.setp(ax11.xaxis.get_majorticklabels(), rotation=45, ha='right')
for i in range(len(tech_labels)):
    for j in range(len(metric_labels)):
        text = ax11.text(j, i, tech_data[i, j],
                        ha="center", va="center", color="black", fontweight='bold')
ax11.set_title('Tech Stack Performance Matrix', fontweight='bold', pad=10)
plt.colorbar(im, ax=ax11, label='Score (out of 100)')

# 12. Ecosystem Flow Diagram (Sankey-style visualization)
ax12 = plt.subplot(3, 4, 12)
stages_flow = ['Developers', 'AI Assist', 'Dojo Build', 'Payment\nSetup', 'Launch']
flow_values = [125, 118, 115, 108, 98]
colors_flow = plt.cm.plasma(np.linspace(0.2, 0.8, len(stages_flow)))

bars = ax12.barh(stages_flow, flow_values, color=colors_flow, alpha=0.8)
ax12.set_xlabel('Active Count', fontweight='bold')
ax12.set_title('Developer Journey Flow', fontweight='bold', pad=10)
ax12.grid(True, alpha=0.3, axis='x')
for i, (bar, val) in enumerate(zip(bars, flow_values)):
    retention = (val/flow_values[0])*100
    ax12.text(val + 2, i, f'{val} ({retention:.0f}%)',
             va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# Print Summary Statistics
print("\n" + "="*80)
print("DOJO GAME LAUNCHPAD - PLATFORM SUMMARY")
print("="*80)
print("\nMOBILE PLATFORM FOR INDIE GAME DEVELOPERS")
print("-" * 80)
print(f"Total Developers Onboarded: {devs_onboarded[-1]}")
print(f"Total Games Launched: {games_launched[-1]}")
print(f"AI-Assisted Launches: {ai_assisted[-1]} ({(ai_assisted[-1]/games_launched[-1]*100):.1f}%)")
print(f"Average Launch Time: {sum(avg_time_hours):.1f} hours")

print("\nMULTI-CHAIN PAYMENT ECOSYSTEM")
print("-" * 80)
print(f"Starknet (Chipi Pay): {payment_volume[0]:.1f}% | {transaction_counts[0]:,} transactions")
print(f"Bitcoin (Xverse): {payment_volume[1]:.1f}% | {transaction_counts[1]:,} transactions")
print(f"Bitcoin (Vesu): {payment_volume[2]:.1f}% | {transaction_counts[2]:,} transactions")
print(f"Total Transaction Volume: {sum(transaction_counts):,}")

print("\nAI AGENT CAPABILITIES")
print("-" * 80)
print(f"Most Used: {ai_features[0]} ({usage_counts[0]} uses)")
print(f"Documentation Generated (RAG): {usage_counts[1]} docs")
print(f"Total AI Interactions: {sum(usage_counts):,}")

print("\nPRIVACY & SECURITY (Wootzapp)")
print("-" * 80)
print(f"Encrypted Chats Active: {privacy_metrics['Active Users'][0]}")
print(f"Average Satisfaction: {privacy_metrics['Satisfaction'].mean():.2f}/5.0")

print("\nOPEN SOURCE COMMUNITY")
print("-" * 80)
print(f"Game Templates Available: {len(templates)}")
print(f"Total Template Downloads: {sum(downloads)}")
print(f"Most Popular: {templates[0]} ({downloads[0]} downloads)")

print("\nPLATFORM REVENUE")
print("-" * 80)
print(f"Total Monthly Revenue: ${sum(monthly_revenue):,}")
print(f"Primary Source: {revenue_sources[0]} (${monthly_revenue[0]:,})")

print("\nKEY TECHNOLOGIES")
print("-" * 80)
print("- Dojo Engine: Game development framework")
print("- AI Agent: RAG-powered documentation & publishing assistance")
print("- Chipi Pay: Starknet payment integration")
print("- Xverse & Vesu: Bitcoin payment options")
print("- Wootzapp: End-to-end encrypted developer communications")
print("- Mobile Launcher: Cross-platform game distribution")
print("="*80)

# Additional visualization: Developer satisfaction over time
fig2, axes = plt.subplots(2, 2, figsize=(16, 10))
fig2.suptitle('Developer Experience & Platform Health Metrics',
              fontsize=18, fontweight='bold')

# Developer Satisfaction Trends
ax_sat = axes[0, 0]
satisfaction_months = months
satisfaction_scores = [4.2, 4.3, 4.5, 4.6, 4.7, 4.8]
ax_sat.plot(satisfaction_months, satisfaction_scores, 'o-', linewidth=3,
            markersize=12, color='green', markerfacecolor='lightgreen')
ax_sat.fill_between(range(len(satisfaction_months)),
                     [s-0.2 for s in satisfaction_scores],
                     [s+0.1 for s in satisfaction_scores],
                     alpha=0.2, color='green')
ax_sat.set_xlabel('Month', fontweight='bold')
ax_sat.set_ylabel('Satisfaction Score (1-5)', fontweight='bold')
ax_sat.set_title('Developer Satisfaction Trend', fontweight='bold', pad=10)
ax_sat.set_ylim([4.0, 5.0])
ax_sat.grid(True, alpha=0.3)

# Support Ticket Resolution
ax_support = axes[0, 1]
support_categories = ['Tech\nSupport', 'Payment\nIssues', 'AI Agent\nHelp', 'Privacy\nQuestions']
tickets_received = [234, 89, 167, 56]
tickets_resolved = [228, 87, 165, 56]
x_support = np.arange(len(support_categories))
width = 0.35
ax_support.bar(x_support - width/2, tickets_received, width,
               label='Received', alpha=0.8, color='orange')
ax_support.bar(x_support + width/2, tickets_resolved, width,
               label='Resolved', alpha=0.8, color='green')
ax_support.set_ylabel('Ticket Count', fontweight='bold')
ax_support.set_title('Support Ticket Resolution', fontweight='bold', pad=10)
ax_support.set_xticks(x_support)
ax_support.set_xticklabels(support_categories)
ax_support.legend()
ax_support.grid(True, alpha=0.3, axis='y')

# Game Categories Distribution
ax_categories = axes[1, 0]
game_categories = ['Strategy', 'RPG', 'Puzzle', 'Action', 'Card', 'Casual']
games_by_category = [18, 15, 12, 9, 5, 2]
colors_cat = plt.cm.Set2(np.linspace(0, 1, len(game_categories)))
wedges, texts, autotexts = ax_categories.pie(games_by_category, labels=game_categories,
                                              autopct='%1.1f%%', colors=colors_cat,
                                              startangle=90)
ax_categories.set_title('Game Categories Distribution', fontweight='bold', pad=10)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# Platform Health Score
ax_health = axes[1, 1]
health_components = ['Uptime', 'API\nResponse', 'Payment\nSuccess', 'AI\nAccuracy', 'User\nSatisfaction']
health_scores = [99.8, 98.5, 97.2, 96.8, 96.0]
colors_health = ['green' if s >= 97 else 'orange' if s >= 95 else 'red' for s in health_scores]
bars = ax_health.barh(health_components, health_scores, color=colors_health, alpha=0.7)
ax_health.set_xlabel('Health Score (%)', fontweight='bold')
ax_health.set_title('Platform Health Metrics', fontweight='bold', pad=10)
ax_health.set_xlim([94, 100])
ax_health.grid(True, alpha=0.3, axis='x')
for i, v in enumerate(health_scores):
    ax_health.text(v - 1, i, f'{v}%', va='center', ha='right',
                   fontweight='bold', color='white')

plt.tight_layout()
plt.show()

print("\nVisualization Complete - All platform metrics displayed")
```
